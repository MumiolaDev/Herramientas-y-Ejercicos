```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
N = 100
x1 = np.linspace(-1,1,N)
def Legendre(n,x):
    leg = sp.legendre(n)
    P_n = leg(x)
    return P_n

for i in range(7):
    f = Legendre(i, x1)
    p1 = plt.plot(x1,f)
```

```python
N = 100
x1 = np.linspace(0,50,N)
def Bessel_J(m,x):
    bes = sp.jv(m,x)
    return bes

for i in range(7):
    f = Bessel_J(i,x1)
    p2 = plt.plot(x1,f)
```

```python
N = 100
x1 = np.linspace(1,5,N)
def Bessel_N(m,x):
    bes = sp.yn(m,x)
    return bes
for i in range(0,7):
    f = Bessel_N(i,x1)
    p3 = plt.plot(x1,f)
```
